```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
data = pd.read_csv('sms_raw_NB.csv', encoding = "ISO-8859-1")
```

```python
data['type'] = data['type'].map({'ham':0,'spam':1})
```

```python
data.head(10)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cc = CountVectorizer(lowercase=True,ngram_range=(1,4))
```

```python
x = cc.fit_transform(data.text)
```

```python
cc.get_feature_names()
```

```python
x.shape
```

```python
from sklearn.model_selection import train_test_split
```

```python
xTrain,xTest, yTrain,yTest = train_test_split(x,data.type,test_size=0.2)
```

```python
from sklearn.naive_bayes import MultinomialNB
```

```python
mnb = MultinomialNB()
```

```python
mnb= mnb.fit(xTrain,yTrain)
```

```python
pred = mnb.predict(xTest)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(yTest,pred)
```

```python
mnb.score(xTest,yTest)
```

```python
mnb.score(xTrain,yTrain)
```

```python
from sklearn.pipeline import Pipeline
```

```python
pipe = Pipeline([
    ('vectorizer',CountVectorizer()),
    ('MNB',MultinomialNB())
])
```

```python
pipe.fit(data.text,data.type)
```
